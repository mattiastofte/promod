```python
from pylab import *
import math
data = loadtxt("temperatur.txt",skiprows=1)

def gjennomsnitt(x):
    i = (1/len(x))*sum(x)
    return i

def standardavvik(x):
    f = 0
    snitt = gjennomsnitt(x)
    for i in x:
        f += (i-snitt)**2
    f = (f*(1/len(x)))**0.5
    return f

a,b,c = polyfit(tid, temp, 2)

def modell(x):
    return a*x**2 + b*x + c

x = linspace(0,400,10000)
y = modell(x)

tid = data[:,0]
temp = data[:,1]

plot(x,y,label="datapunkter")
plot(tid,temp,color='crimson',label="temp",linestyle="",marker=".",)
legend()
xlabel('tid (min)')
ylabel('temperatur (celcius)')

print(gjennomsnitt(temp),"+-",standardavvik(temp))

```

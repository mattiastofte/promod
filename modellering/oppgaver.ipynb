```python
from matplotlib import *
# Oppgave 1
# 1) I_t+1 = ny verdi for antall individer, I_t = antall individer, a = konstant (vekstrate)
# 2) Enkelt modell for eksponensiell vekst (har ikke med begrensninger)
# 3) a beskriver smitteraten, grafen vokser eksponensielt. Når a øker => øker veksten.
# 4) per enhet forklarer fysiskt hvor fort individer vokser.

I = [3]
t = [0]
a = 0.2 # smitterate
uker = 48

for i in range(1,uker+1):
    I.append(I[i-1]+(a*I[i-1]))
    t.append(i)
pyplot.plot(t,I)
```

```python
# Oppgave 2
# 1) Vi tar med hvem som faktisk kan smittet inn i modellen. Dette blir en begrensning.
# 2) Grafen til motakkelige er motsatt av antall individer (i forhold til hverandre).
# 3) Når vi øker a => større vekst, når vi økte antall smittede på begynnelsen "kuttet" vi ut veksten på venstresiden av grafen (på begynnelsen)
# 4) Kan være hensiktmessige når man vi beskrive smitte av sykdom (vi tar med egenskaper og begrensning)

I = [2000]
t = [0]
M = [157756]
populasjon = 157759
S = [(157759-I[0])/157759]
a = 0.2 # smitterate
uker = 48

for i in range(0,uker):
    I.append(I[i]+(a*I[i]*S[i]))
    S.append((populasjon-I[i])/populasjon)
    M.append(populasjon-I[i])
    t.append(i+1)
plt.plot(t,I,label="individer syke")
plt.plot(t,M,label="mottakelige")
plt.xlabel("uker")
plt.ylabel("individer")
plt.legend()
```

```python
# Oppgave 3
# 1) Betydning b: per uke, hvor mange av antall individer (i prosent) blir friske. Halvparten av a. (hvis man ønsker at man blir frisk på 2 uker)
# 2) 

I = [2000]
t = [0]
M = [157756]
P = [157759]
R = [0]
S = [(P[0]-I[0])/P[0]]
a = 0.4 # rate for antall som blir syke
b = 0.1 # rate for antall som blir friske
uker = 48

for count in range(0,uker):
    I.append(I[count]+(a*I[count]*S[count])-b*I[count]) # oppdater
    S.append((populasjon-I[count]-R[count])/populasjon)
    R.append(R[count]+b*I[count])
    M.append(populasjon-I[count]-R[count])
    t.append(count+1)
plt.plot(t,I,label="individer syke")
plt.plot(t,M,label="mottakelige")
plt.plot(t,R,label="immune")
plt.plot(t,)
plt.xlabel("uker")
plt.ylabel("individer")
plt.legend()
```

```python
# rater
a = 0.4 # for antall som blir syke
b = 0.1 # antall som blir friske
d = 0.023 # antall som dør av sykdommen
vaksine_effektivitet = 0.1
uker = 48

I = [2000]
t = [0] # motakkelige
P = [157759] 
R = [P[0]*vaksine_effektivitet]
M = [157756-P[0]*vaksine_effektivitet]
S = [(P[0]-I[0])/P[0]]

for count in range(0,uker):
    I.append(I[count]+(a*I[count]*S[count])-b*I[count]) # oppdater, syke
    S.append(M[count]/P[count]) # rate med hensyn på "smittbare"
    R.append(R[count]+(b-d)*I[count]) # resistente
    M.append(P[count]-I[count+1]-R[count]) # mottakelige
    P.append(P[count]-d*I[count]) 
    t.append(count+1)
plot(t,I,label="individer syke")
plot(t,M,label="mottakelige")
plot(t,R,label="resistente")
plot(t,P,label="populasjon")
xlabel("uker")
ylabel("individer")
legend()
print("antall døde: ",P[0]-P[-1])

# Konklusjon: vaksine = good
```

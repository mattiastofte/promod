## Numerisk derivasjon
Vi tilnærmer grenseverdien med en liten $ \Delta x $

$ f(x)=\frac{f(x+\Delta x)-f(x)}{\Delta x} $

```python
def f(x):
    return x**2

def derivert(f, x, dx):
    derivert =  (f(x + dx) - f(x)) / dx
    return derivert

# LØKKE FOR Å GENERERE LISTE MED STANDARDFORMER
verdier = []
for n in range(16):
    verdier.append(float(f'1E-{n+1}'))

for standardform in verdier:
    print(f'{standardform}: har en feil på {round(abs(2-derivert(f,1, standardform)),15)}')

```

## Diskret derivasjon
- Vi kan også derivere diskrete data, ikke bare kontinuerlige funksjoner.
- Dette kan vi ikke gjøre med analytiske metoder.

```python
from pylab import *
from numpy import *
from matplotlib.pyplot import *
data = loadtxt('heistur.csv', skiprows = 1, delimiter = ',')
```

```python
tid = data[:,0]
posisjon = data[:,2]

plot(tid,posisjon)
```

```python
fart = []
for i in range(len(tid)-1):
    dy = posisjon[i+1] - posisjon[i]
    dx = tid[i+1] - tid[i]
    derivert = dy/dx
    fart.append(derivert)

plot(tid[:-1], fart)
```

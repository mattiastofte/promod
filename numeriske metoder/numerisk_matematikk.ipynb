```python
import math
g = lambda x: -4*x+4

def solve(a,b,f):
    while not round(f(b),6)==0:
        delta = (b-a)/2
        if f(b-delta) > 0:
            a = b-delta
        else:
            b = b-delta
    return round(b,6)
print(solve(-10000,10000,g))
```

```python
import math
i = lambda x: x**4 - 10

def solve(a,b,f,tol=1E-8):
    m = (a+b)/2
    while abs(f(m)) >= tol:
        if f(a)*f(m) < 0:
            b = m
        elif f(b)*f(m) < 0:
            a = m
        m = (a+b)/2
    return m

print(solve(-50,50,i))
```

```python
a = 0        # Startverdi
b = 5        # Sluttverdi
dx = 1E-5    # Steglengde
N = int((b-a)/dx) # Antall intervaller
y0 = 1       # Initialbetingelse

f_der = lambda x,y: 1
```

```python
import numpy as np

x = np.zeros(N+1)
y = np.zeros(N+1)
```

Vi lager arrays med N+1 ettersom N representerer antall intervaller, og dette tallet er 1 mindre enn antall punkter (x, y)

```python
# Eulers metode

import matplotlib.pyplot as plt
y[0] = y0

def solve(f_der,x,y):
    for i in range(N):
        y[i+1] = y[i] + f_der(x[i], y[i])*dx
        x[i+1] = x[i] + dx

solve(f_der,x,y)
plt.plot(x,y)
```

```python
x_start = 0
x_slutt = 10
T_omg = 21
dx = 1E-5
N = int((x_slutt-x_start)/dx)
k = 0.4
T0 = 100

N_avkjøling = lambda T: -k*(T-T_omg)

t = np.zeros(N+1)
T = np.zeros(N+1)

T[0] = T0

for n in range(N):
    T[n+1] = T[n] + N_avkjøling(T[n])*dx
    t[n+1] = t[n] + dx

plt.plot(t,T)
```

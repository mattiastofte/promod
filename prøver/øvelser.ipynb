```python
import numpy as np
from pylab import *
from matplotlib.pyplot import *
import math

N = 10000
x_0 = 0
x_1 = 10
h = (x_1-x_0)/(N-1)

x = np.zeros(N)
y = np.zeros(N)

y[0] = 0

def y_der(x,y):
    return math.sin(x) - y**3

for i in range(N-1):
    y[i + 1] = y[i] + y_der(x[i],y[i])*h
    x[i + 1] = x[i] + h

plot(x,y)
```

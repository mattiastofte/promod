```python
# OPPGAVE 4

def delelige_heltall(a,b,c):
    f = []
    for i in range(a,b):
        if i%c == 0:
            f.append(i)
    return f

print("mellom 0 og 1000 finnes det",len(delelige_heltall(0,1000,7)),"delelige heltall som kan deles på 7")
```
